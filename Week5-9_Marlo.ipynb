```python
import networkx as nx
import itertools
from itertools import chain

import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Sample from Gaussian Distribution
N = 10000
mu = 0
sigma = 4
X = np.random.standard_normal(N) * sigma + mu

# Create Figure
plt.figure(figsize=(10,6))

# 2. Plot Distribution
plt.hist(X)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Sampled Data from Gaussian Distribution')

# 3. Cumulative Avg
def get_cumulative_average(data):
    cumulative_av = np.array([(sum(data[:i+1])/(i+1)) for i in range(len(data))])
    return cumulative_av

# 4. Cumulative Std Error
def get_cumulative_error(data):
    n = len(data)

    stdev_array = np.zeros_like(data)
    for i in range(len(data)):
        stdev_array[i] = np.std(data[:i+1])
    
    cumulative_error = stdev_array/np.sqrt(n)
    return cumulative_error

cumulative_average = get_cumulative_average(X)
cumulative_std_error = get_cumulative_error(X)

# 5. Distribution mean & median
distribution_median = np.median(X)
distribution_mean = np.mean(X)

# 6. Plot Cumulative Avg with error bars & mean line
plt.figure(figsize=(10, 6))
x = np.arange(1, N+1)
plt.errorbar(x[::10], cumulative_average[::10], yerr=cumulative_std_error[::10],fmt='-o', color='b', label='Cumulative Average')
plt.plot(distribution_mean)
plt.ylabel('Cumulative Average')
plt.xlabel('Number of samples')
plt.title('Cumulative Avg (with Error Bars) ')
plt.legend()

plt.show()
```
