```python
def lex_diversity(text): 
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total
```

### Part 3 - Putting things into practice with abstract dataset 

```python
def convert_abstract_to_text(row):
    text = ""
    for key, val in row.items():
        print(key)
        text = text + key
    return text 
```

```python
import json
```

```python
def convert_abs_to_dict(s):
    print(s)
    s = s.replace("'", "\"")
    dict_from_string = json.loads(s)
    return dict_from_string
```

```python
abstracts_df = pd.read_csv("data/final_arbstract.csv")
abstracts_df = abstracts_df.drop(columns =["Unnamed: 0"])

abstracts_df["abstract_inverted_index"] = abstracts_df["abstract_inverted_index"].apply(convert_abs_to_dict)
abstracts_df["abstract_inverted_index"] = abstracts_df["abstract_inverted_index"].apply(convert_abstract_to_text)
```
